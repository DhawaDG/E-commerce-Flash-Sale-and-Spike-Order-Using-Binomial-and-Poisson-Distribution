#### Convert date columns to datetime

```python
daily_df['date']=pd.to_datetime(daily_df['date'])
daily_df.info()
daily_df.head()
```

```python
hourly_df['date']=pd.to_datetime(daily_df['date'])
hourly_df.info()
hourly_df.head()
```

#### Check for dublicacy rows

```python
daily_duplicates = daily_df.duplicated().sum()

print(f'Number of duplicates  row in Flash sale daily : {daily_duplicates}')
```

```python
hourly_duplicates = hourly_df.duplicated().sum()

print(f'Number of duplicates  row in Flash sale hourly : {hourly_duplicates}')
```

#### Distribution plots for daily data

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(16, 10))
for i, col in enumerate(['visitors', 'customers', 'conversion_rate', 'orders', 'orders_per_customer', 'aov', 'revenue']):
    plt.subplot(3, 3, i+1)
    sns.histplot(daily_df[col], kde=True, bins=20)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()
```

#### Distribution plots for hourly data

```python
plt.figure(figsize=(16, 8))
for i, col in enumerate(['visitors', 'customers', 'conversion_rate', 'orders', 'orders_per_customer']):
    plt.subplot(2, 3, i+1)
    sns.histplot(hourly_df[col], kde=True, bins=20)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()
```
